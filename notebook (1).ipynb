## üìñ Background
You volunteer for a public policy advocacy organization in Canada, and your colleague asked you to help her draft recommendations for guidelines on CO2 emissions rules. 

After researching emissions data for a wide range of Canadian vehicles, she would like you to investigate which vehicles produce lower emissions.

## üíæ The data I

### You have access to seven years of CO2 emissions data for Canadian vehicles ([source](https://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64#wb-auto-6)):

- "Make" - The company that manufactures the vehicle.
- "Model" - The vehicle's model.
- "Vehicle Class" - Vehicle class by utility, capacity, and weight.
- "Engine Size(L)" - The engine's displacement in liters.
- "Cylinders" - The number of cylinders.
- "Transmission" - The transmission type: A = Automatic, AM = Automatic Manual, AS = Automatic with select shift, AV = Continuously variable, M = Manual, 3 - 10 = the number of gears.
- "Fuel Type" - The fuel type: X = Regular gasoline, Z = Premium gasoline, D = Diesel, E = Ethanol (E85), N = natural gas.
- "Fuel Consumption Comb (L/100 km)" - Combined city/highway (55%/45%) fuel consumption in liters per 100 km (L/100 km).
- "CO2 Emissions(g/km)" - The tailpipe carbon dioxide emissions in grams per kilometer for combined city and highway driving. 

The data comes from the Government of Canada's open data [website](https://open.canada.ca/en).

```python
# Import the pandas and numpy packages
import pandas as pd
import numpy as np

# Load the data
cars = pd.read_csv('data/co2_emissions_canada.csv')

# create numpy arrays
cars_makes = cars['Make'].to_numpy()
cars_models = cars['Model'].to_numpy()
cars_classes = cars['Vehicle Class'].to_numpy()
cars_engine_sizes = cars['Engine Size(L)'].to_numpy()
cars_cylinders = cars['Cylinders'].to_numpy()
cars_transmissions = cars['Transmission'].to_numpy()
cars_fuel_types = cars['Fuel Type'].to_numpy()
cars_fuel_consumption = cars['Fuel Consumption Comb (L/100 km)'].to_numpy()
cars_co2_emissions = cars['CO2 Emissions(g/km)'].to_numpy()

# Preview the dataframe
cars
```

## üí™ Challenge I
Help your colleague gain insights on the type of vehicles that have lower CO2 emissions. Include:

1. What is the median engine size in liters?
2. What is the average fuel consumption for regular gasoline (Fuel Type = X), premium gasoline (Z), ethanol (E), and diesel (D)?  
3. What is the correlation between fuel consumption and CO2 emissions?
4. Which vehicle class has lower average CO2 emissions, 'SUV - SMALL' or 'MID-SIZE'? 
5. What are the average CO2 emissions for all vehicles? For vehicles with an engine size of 2.0 liters or smaller?
6. Any other insights you found during your analysis?

```python
#1.What is the median engine size in liters?
cars['Engine Size(L)'].median()
```

```python
#2.What is the average fuel consumption for regular gasoline (Fuel Type = X), premium gasoline (Z), ethanol (E), and diesel (D)?

#let's create a dictionary of fuel names to recall names in final answer
fuel= {'X':'gasoline','Z':'premium gasoline','E':'ethanol','D':'diesel'}

#let's group by data by fuel type & average out fuel consumption by fuel type           
avg_fuel_consumption= cars.groupby(['Fuel Type']).agg({'Fuel Consumption Comb (L/100 km)':'mean'}).reset_index()

#let's use 'fuel' dictionary for fuel names & use map function to get fuel names as a column
avg_fuel_consumption['Fuel Type']=avg_fuel_consumption['Fuel Type'].map(fuel)


#as the Fuel Type column in cars DataFrame as more types than listed for this question, there can be null values crated in the output below. Hence, print the output & drop null values
print(avg_fuel_consumption.dropna())
```

```python
#3.What is the correlation between fuel consumption and CO2 emissions?
corr= cars['Fuel Consumption Comb (L/100 km)'].corr(cars['CO2 Emissions(g/km)'])
print(corr)

#these two columns shows strong positive correlation of 0.91.
```

```python
#4. Which vehicle class has lower average CO2 emissions, 'SUV - SMALL' or 'MID-SIZE'?

#Let's create a temporary DataFrame which contains only two vehicle class 
temp=cars.loc[(cars['Vehicle Class']=='SUV - SMALL') | (cars['Vehicle Class']=='MID-SIZE'), :]

#Let's average out CO2 emissions by Vehicle Class column
SUV_small= temp.groupby(['Vehicle Class']).agg({'CO2 Emissions(g/km)':'mean'})

#Print final result
print(SUV_small)

#MID_SIZE vehicle class has lower CO2 emissions, i.e. 222.45g/km
```

```python
#5.What are the average CO2 emissions for all vehicles? For vehicles with an engine size of 2.0 liters or smaller?

# Part I :average CO2 emissions for all vehicles? 

#average out CO2 emissions by vehicles class
avg_emissions= cars.groupby(['Vehicle Class']).agg({'CO2 Emissions(g/km)':'mean'})

#print result
print(avg_emissions)

# Part II :average CO2 emissions for vehicles for vehicles with an engine size of 2.0 liters or smaller?

#Create a dataframe of vehicles with an engine size of 2.0 liters or smaller
engine=cars['Engine Size(L)']<=2.0
engine=cars[engine]

#average out CO2 emissions by vehicles class
avg_emissions_for_engine_size= engine.groupby(['Vehicle Class']).agg({'CO2 Emissions(g/km)':'mean'})

#print result
print(avg_emissions_for_engine_size)
```

```python
#6.Any other insights you found during your analysis?
#Lets look at CO2 emission distribution to understand,where does the emission for most of the vehicle modle lies?

import matplotlib.pyplot as plt
Co2_emissions = cars['CO2 Emissions(g/km)']

#Let's plot histagram & set bins = 20
plt.hist(Co2_emissions, bins= 20)
plt.plot()
plt.show()

#Maximum car models falls in the emission range of 200 to 300 g/kg CO2 emission
```

## üí™ Challenge II
Help your team leader understand your company's products. Include:

1. What is the most expensive item your company sells? The least expensive?
2. How many different products of each category does your company sell?
3. What are the top three brands with the highest average list price? The top three categories?
4. Any other insights you found during your analysis?

```python
---1. What is the most expensive item your company sells? The least expensive?

--expensive item
SELECT product_name, list_price
FROM products
Where list_price= (select MAX(list_price) from products)

--least expensive
union all
SELECT product_name, list_price
FROM products
Where list_price= (select min(list_price) from products)
;

--The most expensive product is Trek Domane SLR 9 Disc - 2018 at 11999.99 price & the least expensive product is Strider Classic 12 Balance Bike - 2018 priced 89.99.
```

```python
--2. How many different products of each category does your company sell?

Select categories.category_name, count(products.product_id) as product_count

From products
join categories
on products.category_id=categories.category_id 

Group by categories.category_name;
```

```python
--3. What are the top three brands with the highest average list price? The top three categories?

--top three brands
select top 3 brands.brand_name, avg(products.list_price) as avg_price

from products
join brands
on brands.brand_id=products.brand_id

Group by brands.brand_name
order by avg_price Desc;
```

```python
--3. top three categories?
select top 3 categories.category_id, avg(products.list_price) as avg_price

from products
join categories
on categories.category_id=products.category_id

Group by categories.category_id
order by avg_price Desc;
```

```python
--4.Any other insights you found during your analysis?
--Which is the highest sold brand

SELECT brand_id, count(brand_id) as maximum_sold
FROM products
group by brand_id
order by maximum_sold desc
;

--Brand Id 9, that is Trek is the most sold brand in the shop
```
